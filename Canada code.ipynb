```python
import pandas as pd
import requests
import pprint
import matplotlib.pyplot as plt
```

```python
CA_df = pd.read_csv("Video_CSVs/CAvideos.csv")
CA_df.head()
params = {"part":"snippet","regionCode":"US",
           "key":"AIzaSyBPeLWHKQa9WjkzYQMRgrSVTsH1O4gtMIE"}
base_url= "https://www.googleapis.com/youtube/v3/videoCategories"
response = requests.get(base_url, params=params).json()
```

```python
id_list = []
title_list = []
for i in response['items']:
    #print(i)
    id_list.append(i['id'])
    title_list.append(i['snippet']['title'])
# Create DF
new_df=pd.DataFrame({"category_id":id_list,"title_list":title_list})
new_df["category_id"]=new_df["category_id"].astype("int64")
new_df.dtypes
# Merge with us_df
CA_df=pd.merge(CA_df,new_df,on='category_id',how="left")
CA_df.drop(columns=["publish_time","tags","thumbnail_link","comments_disabled","ratings_disabled","video_error_or_removed","description"],inplace=True)
CA_df
```

```python
#Renaming the Columns
CA_df_RN=CA_df.rename(columns={"video_id":"Video ID", "trending_date": "Trending Date",'title': "Title",
                        "channel_title":"Channel Title",
                       'category_id': "Category ID", 'views':'Views', "likes": 'Likes', 'dislikes': "Dislikes",
                        'comment_count': 'Comment Count'})
```

```python
# Find top viewed Channels
CA_channel_data = CA_df_RN.loc[:,("Channel Title", "Views", "Likes", "Dislikes")]
CA_channel_df = CA_channel_data.head(5)
CA_channel_df
```

```python
# sum the views, likes, and dislikes for each channel
CA_topviewedchannel = CA_channel_df.groupby("Channel Title").sum()["Views"]
CA_toplikedchannel = CA_channel_df.groupby("Channel Title").sum()["Likes"]
CA_topdislikedchannel = CA_channel_df.groupby("Channel Title").sum()["Dislikes"]

# create a dataframe of the total views, likes, and dislilkes of each channel
CA_channel_summary = pd.DataFrame({"Total Views":CA_topviewedchannel,
                               "Total Likes":CA_toplikedchannel,
                               "Total Dislikes":CA_topdislikedchannel})

# print dataframe
CA_channel_summary_df = CA_channel_summary.head(5)
CA_channel_summary_df = CA_channel_summary_df.reset_index()
CA_channel_summary_df
```

```python

labels = CA_channel_summary_df["Channel Title"]
width = 0.35 

fig, ax = plt.subplots()

plt.bar(labels, CA_channel_summary_df["Total Views"], width, label='Views')
#plt.bar(labels, CA_channel_summary_df["Total Dislikes"], width, label='Dislikes')
plt.bar(labels, CA_channel_summary_df["Total Likes"], width, label='Likes')

plt.ylabel('Totals')
plt.title('Channel Title')
plt.legend()

plt.show()
```

```python
explode = (0.1, 0, 0, 0, 0)
colors = ["springgreen", "lightseagreen", "dodgerblue", "mediumorchid", "hotpink"]
plt.pie(CA_channel_summary_df["Total Views"], explode=explode, labels=CA_channel_summary_df["Total Views"], colors = colors, autopct="%1.1f%%", shadow=True)
plt.title("Total top 5 Viewed YouTube Channels")
```
